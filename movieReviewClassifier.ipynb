```python
import pandas as pd
import numpy as np
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from string import punctuation
from nltk.stem import WordNetLemmatizer
import random
from nltk import NaiveBayesClassifier, classify
```

```python
df = pd.read_csv(f'/Users/shashwat/Documents/Local Disk D/ML Datasets/IMDB Dataset.csv')
df.head(3)
```

```python
l = WordNetLemmatizer()
```

```python
def clean_text(text):
    words = word_tokenize(str(text).lower())
    stopword = stopwords.words("english")
    stopword = list(punctuation) + stopword
    words = [word for word in words if word.lower() not in stopword]
    words = [l.lemmatize(word) for word in words]
    return words
    
def get_features(words):
    return {word: True for word in words}
```

```python
dataset = [(get_features(clean_text(row["review"])), row["sentiment"])
           for _, row in df.iterrows()]
```

```python
random.shuffle(dataset)
train_data = dataset[:40000]
test_data = dataset[40000:]
```

```python
classifier = NaiveBayesClassifier.train(train_data)
accuracy = classify.accuracy(classifier, test_data)*100
print(f"Model Accuracy:{accuracy:.2f}%")
classifier.show_most_informative_features(10)
```

```python
test_review = "I loved the movie very much"
features = get_features(clean_text(test_review))
print("Sentiment:", classifier.classify(features))
```
